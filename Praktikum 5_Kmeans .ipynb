```python
#202031265_Yeni Anggraini Fadhilah
data=pd.read_csv('creditcard.csv')
data
```

```python
#202031265_Yeni Anggraini Fadhilah
data=data[['BALANCE','PURCHASES']]
data
```

# Persebaran data dengan scatter

```python
#202031265_Yeni Anggraini Fadhilah
plt.scatter(data.BALANCE, data.PURCHASES)
plt.show()
```

# Persebaran data dengan boxplot

```python
#202031265_Yeni Anggraini Fadhilah
kolom=['BALANCE','PURCHASES']
plt.boxplot(data[kolom])
plt.xticks([1,2],kolom)
plt.title('Outlier Sebelum dibuang')
plt.show
```

# Menghapus data (data outlayer) menggunakan IQR 

```python
#202031265_Yeni Anggraini Fadhilah
Q1= data[kolom].quantile(0.25)
Q3= data[kolom].quantile(0.75)
IQR= Q3-Q1
min_iqr =(Q1-1.5 *IQR)
max_iqr =(Q3+1.5 *IQR)

data= data[~((data[kolom]<min_iqr) | (data[kolom]>max_iqr)).any(axis=1)]

plt.boxplot(data[kolom])
plt.xticks([1,2],kolom)
plt.title('Outlier Setelah dibuang')
plt.show
```

```python
#202031265_Yeni Anggraini Fadhilah
plt.scatter(data.BALANCE, data.PURCHASES)
plt.show()
```

```python
#202031265_Yeni Anggraini Fadhilah
data.describe()
```

#  Melakukan "Scaling" agar jarak nilai (std) tidak terlalu jauh

```python
#202031265_Yeni Anggraini Fadhilah
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_array= np.array(data)
x_scaled= scaler.fit_transform(x_array)

plt.scatter(x_scaled[:,0], x_scaled[:,1])
```

# Membuat model "clustering"

```python
#202031265_Yeni Anggraini Fadhilah
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3, random_state=0)
kmeans.fit(x_scaled)
```

# Mencari nilai centroid/Center/Titik tengah

```python
#202031265_Yeni Anggraini Fadhilah
kmeans.cluster_centers_
```

# Menampilkan grafiknya

```python
#202031265_Yeni Anggraini Fadhilah
output = plt.scatter(x_scaled[:,0], x_scaled[:,1],  s=10, c=kmeans.labels_)
centers = kmeans.cluster_centers_

plt.scatter(centers[:,0], centers[:,1], c='red', s=10)
plt.title("KMeans Clustering")
plt.xlabel("Balance")
plt.ylabel("Purchases")
plt.colorbar(output)
plt.show()
```

# Model Evaluation

```python
#202031265_Yeni Anggraini Fadhilah
from sklearn.metrics import davies_bouldin_score

labels= kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```
